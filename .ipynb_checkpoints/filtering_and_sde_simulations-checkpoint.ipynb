## SDE simulations

Time specification

```julia
Δₜ = 1 / 1000
time = (1:1000) / 1000;
```

Simulating standard brownian motion

```julia
dW = [0.0]
append!(dW, rand(Normal(0, Δₜ ^ .5), length(time) - 1))
W = cumsum(dW)
plot(time, W)
```

Solution to SDE: $$dX_{t}=\frac{\exp(W_{t})}{1+\exp(W_{t})}dW_{t}, X_{0}=0$$

```julia
X = cumsum(cdf.(Logistic(), W) .* dW)
plot(time, X)
```

Solution to SDE: $$dX_{t}=\frac{\exp(X_{t})}{1+\exp(X_{t})}dW_{t}, X_{0}=0$$

```julia
X = [0.0]
for k in 2:length(time)
    push!(X, X[end] + cdf.(Logistic(), X[end]) * dW[k])
end # end for
plot(time, X)
```

Solution to: $$dX_{t}=cos(W_{t})dW_{t}+sin(X_{t})dt, X_{0}=0$$

```julia
X = [0.0]
for k in 2:length(time)
    push!(X, X[end] + cos(W[k-1]) * dW[k] + sin(X[end]) * Δₜ)
end # end for
plot(time, X)
```

Solution to: $$dX_{t}=cos(X_{t})dW_{t}+sin(X_{t})dt, X_{0}=0$$

### Euler–Maruyama method

```julia
X = [0.0]
for k in 2:length(time)
    push!(X, X[end] + cos(X[end]) * dW[k] + sin(X[end]) * Δₜ)
end # end for
plot(time, X)
```

### Runge–Kutta method

```julia
X = [0.0]
for k in 2:length(time)
    temp = X[end] + sin(X[end]) * Δₜ + cos(X[end]) * sqrt(Δₜ)
    push!(X, X[end] + cos(X[end]) * dW[k] + sin(X[end]) * Δₜ + .5 * (Δₜ ^ -.5) * (cos(temp) - cos(X[end])) * (dW[k] ^ 2 - Δₜ))
end # end for
plot(time, X)
```

Difference between two methods is small

```julia
X1 = [0.0]
for k in 2:length(time)
    push!(X1, X1[end] + cos(X1[end]) * dW[k] + sin(X1[end]) * Δₜ)
end # end for
plot(time, X - X1)
```

## Filtering

### Drift detection

Suppose that we observe stochastic process $\xi = (\xi_{t}, \mathcal{F}_{t})$ which satisfies:
$$d\xi_{t}=\theta dt+dW_{t}$$
where $\theta\sim b(p)$. Then an optimal estimator, in terms of mean squared error, for $\theta$ given $\xi$ at time $t$ is:
$$\hat{\theta}^{(t)}=\mathbb{E}\left[\theta|\mathcal{F}_{t}^{\xi}\right]=\frac{\frac{p}{1-p}\exp\left(\xi_{t}-\frac{t}{2}\right)}{1+\frac{p}{1-p}\exp\left(\xi_{t}-\frac{t}{2}\right)}$$
which also satisfies SDE:
$$d\hat{\theta}^{(t)}=\hat{\theta}^{(t)}\left(1-\hat{\theta}^{(t)}\right)d\xi_{t}-\left(\hat{\theta}^{(t)}\right)^{2}\left(1-\hat{\theta}^{(t)}\right)dt, \hat{\theta}^{(0)}=p$$

```julia
dW = [0.0]
append!(dW, rand(Normal(0, Δₜ ^ .5), length(time) - 1))
W = cumsum(dW)
p = .5
θ = rand(Bernoulli(p))

dξ = [0.0]
append!(dξ, dW[2:end] .+ θ * Δₜ)
ξ = cumsum(dξ)
θ̂ = cdf.(Logistic(), (p / (1 - p)) .* exp.(ξ .- time ./ 2))
plot(time, θ̂)
```

```julia
plot(time, [θ̂, ξ], label=["θ̂" "ξ"])
xlabel!("Time|θ = " * string(θ * 1) * "|Last estimate = " * string(round(θ̂[end]; digits = 3)))
ylabel!("Values")
```

So knowledge of p is crucial.

When number of observations is greater than 1

```julia
Δₜ = 1 / 1000
time = (1:1000) / 1000
n = 5

dW = Array{Real}(undef, (1000, n))
W  = Array{Real}(undef, (1000, n))
dW[1, :] = zeros(n)
for k in 2:length(time)
    dW[k, :] = rand(Normal(0, Δₜ ^ .5), n)
end # end for

# Creating W from dW
for k in 1:n
    W[:, k] = cumsum(dW[:, k])
end # end for
p = .5
θ = rand(Bernoulli(p))

dξ = Array{Real}(undef, (1000, n))
ξ  = Array{Real}(undef, (1000, n + 1))
dξ[1, :] = zeros(n)
dξ[2:end, :] = dW[2:end, :] .+ θ * Δₜ

for k in 1:n
    ξ[:, k] = cumsum(dξ[:, k])
end # end for

ξ[:, n+1] = map(mean, eachrow(ξ[:, 1:n]))


plot(time, ξ, 
    label = reduce(hcat, [permutedims(["ξ_$k" for k in 1:n]), "mean"]), 
    linewidth = reduce(vcat, [[1 for k in 1:n], 3.5])')
xlabel!("Time")
ylabel!("ξ values")
```

```julia
θ̂ = Array{Real}(undef, (1000, n + 1))
θ̂[:, 1:n] = cdf.(Logistic(), (p / (1 - p)) .* exp.(ξ[:, 1:n] .- time ./ 2))
θ̂[:, n+1] = map(mean, eachrow(θ̂[:, 1:n]))

plot(time, θ̂, 
    label = reduce(hcat, [permutedims(["θ̂_$k" for k in 1:n]), "mean"]),
    linewidth = reduce(vcat, [[1 for k in 1:n], 3.5])')
xlabel!("Time|θ = $(1 * θ)" * "|Last estimate = $(round(θ̂[size(θ̂, 1), n+1]; digits = 3))")
ylabel!("ξ values")
```

## Filtering continous time markov chains

### Basic example: error detection

Suppose that we observe stochastic process $\xi = (\xi_{t}, \mathcal{F}_{t})$ which satisfies:
$$d\xi_{t}=\theta_{t} dt+dW_{t}, \xi_{0}=0$$
where $\theta$ is a markov chain starting at $\theta_{0}\sim b(p)$, taking values in $\{0,1\}$ and with transition densities given in matrix:
$$\Lambda=\begin{pmatrix}
-\lambda & \lambda \\
0 & 0
\end{pmatrix}$$
for some $\lambda \in \mathbb{R}_{\geq0}$, i.e elements of matrix $\Lambda$ satisfy:
$$\forall t\in[0,\infty):\left|\mathbb{P}[\theta_{t+\Delta t}=b-1|\theta_{t}=a-1]-\delta_{ab}-\Lambda_{ab}\Delta t\right|\leq o(\Delta t)$$
where $o(\Delta t)/\Delta t\rightarrow0$ when $\Delta t\rightarrow0$, $\delta$ is kronecker delta and $a,b\in\{1,2\}$.

Then an optimal estimator, in terms of mean squared error, for $\mathbb{P}[\theta_{t}=1]$ (i.e. error is present at time $t$ which is equivalent to error occuring before or at time $t$) given $\xi$ $\left(\pi(t)=\mathbb{P}\left[\theta_{t}=1|\mathcal{F}_{t}^{\xi}\right]\right)$ at time $t$ is a solution to SDE:
$$d\pi(t)=\left(\lambda(1-\pi(t))-\pi(t)^{2}(1-\pi(t))\right)dt+\pi(t)(1-\pi(t))d\xi_{t}$$
if we let $\tau$ be defined as the time at which the first error first occured:
$$\tau:=\inf_{t}\mathbb{1}_{\{\theta_{t}=1\}}$$
then $\pi(t)=\mathbb{P}\left[\tau\leq t|\mathcal{F}_{t}^{\xi}\right]$.

```julia
dW = [0.0]
append!(dW, rand(Normal(0, Δₜ ^ .5), length(time) - 1))
W = cumsum(dW)
p = .1
θ = rand(Bernoulli(p))
λ = .5
τ = (θ == 1) ? 0 : Inf

ξ = [0.0]
for k in 2:length(time)
    append!(ξ, ξ[end] + θ * Δₜ + dW[k])
    
    if θ == 0
        θ = rand(Bernoulli(λ * Δₜ))
        if θ == 1
            τ = k
        end # end if
    end # end if
end # end for
τ
```

```julia
dξ = [0.0]
append!(dξ, diff(ξ))

est = [p]
for k in 2:length(time)
    append!(est, est[end] + (λ * (1 - est[end]) - est[end] ^ 2 * (1 - est[end])) * Δₜ + est[end] * (1 - est[end]) * dξ[k])
end # end for
```

```julia
xx = (findfirst(isequal(1), est .> .5) == nothing) ? "No error classification occured" : 
"First error classification at time $(round(findfirst(isequal(1), est .> .5) / 1000; digits = 4))"

plot(time, [est, ξ], label = ["π" "ξ"])
xlabel!("Time|Last estimate = $(round(est[end]; digits = 3))|τ = $(round(τ / 1000; digits = 4))\n" * "$(xx)")
```

Longer observation time

```julia
dξ = [0.0]
append!(dξ, diff(ξ))

est = [p]
for k in 2:length(time)
    append!(est, est[end] + (λ * (1 - est[end]) - est[end] ^ 2 * (1 - est[end])) * Δₜ + est[end] * (1 - est[end]) * dξ[k])
end # end for

xx = (findfirst(isequal(1), est .> .5) == nothing) ? "No error classification occured" : 
"First error classification at time $(round(findfirst(isequal(1), est .> .5) * Δₜ; digits = 4))"

plot(time, [est, ξ], label = ["π" "ξ"])
xlabel!("Time|Last estimate = $(round(est[end]; digits = 3))|τ = $(round(τ * Δₜ; digits = 4))\n" * "$(xx)")
```

More observations

```julia
Δₜ = 1 / 500
time = (1:1000) / 500
n = 10

dW = Array{Real}(undef, (length(time), n))
W  = Array{Real}(undef, (length(time), n))
dW[1, :] = zeros(n)
for k in 2:length(time)
    dW[k, :] = rand(Normal(0, Δₜ ^ .5), n)
end # end for
for k in 1:n
    W[:, k] = cumsum(dW[:, k])
end # end for

p = .01
θ = rand(Bernoulli(p))
λ = .4
τ = (θ == 1) ? 0 : Inf

dξ = Array{Real}(undef, (length(time), n))
ξ  = Array{Real}(undef, (length(time), n + 1))
dξ[1, :] = zeros(n)
ξ[1, 1:n] = zeros(n)


for k in 2:1000
    dξ[k, :] = dW[k, :] .+ θ * Δₜ
    ξ[k, 1:n] = ξ[k-1, 1:n] + dξ[k, :]
    
    if θ == 0
        θ = rand(Bernoulli(λ * Δₜ))
        if θ == 1
            τ = k
        end # end if
    end # end if
end # end for

τ
```

```julia
ξ[:, n+1] = map(mean, eachrow(ξ[:, 1:n]))


plot(time, ξ, 
     label = reduce(hcat, [permutedims(["ξ_$k" for k in 1:n]), "mean"]), 
     linewidth = reduce(vcat, [[1 for k in 1:n], 3.5])')
xlabel!("Time")
ylabel!("ξ values")
```

```julia
est = Array{Real}(undef, (length(time), n + 1))

est[1, 1:n] = zeros(n) .+ p
for k in 2:length(time)
    est[k, 1:n] = est[k-1, 1:n] .+ (λ .* (1 .- est[k-1, 1:n]) .- est[k-1, 1:n] .^ 2 .* (1 .- est[k-1, 1:n])) .* Δₜ .+ est[k-1, 1:n] .* (1 .- est[k-1, 1:n]) .* dξ[k, :]
end # end for


est[:, n+1] = map(mean, eachrow(est[:, 1:n]))

xx = findfirst(isequal(1), est[:, n+1] .> .5)
xx = (xx == nothing) ? "No error classification occured" : 
"First error classification at time $(round(xx * Δₜ; digits = 4))"

plot(time, est, 
    label = reduce(hcat, [permutedims(["π_$k" for k in 1:n]), "mean"]),
    linewidth = reduce(vcat, [[1 for k in 1:n], 3.5])')
xlabel!("Time|Last estimate = $(round(est[size(est)[1], n+1]; digits = 3))|τ = $(round(τ * Δₜ; digits = 4))\n" * "$(xx)")
ylabel!("Values")
```
