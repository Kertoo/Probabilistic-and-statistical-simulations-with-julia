```julia
function generate_brownian_motion(Δₜ, t)
    cumsum(rand(Normal(0, Δₜ ^ .5), round(Int, t / Δₜ)))
end # end function

function generate_brownian_bridge(Δₜ, t)
    x = generate_brownian_motion(Δₜ, t)
    interval = Δₜ:Δₜ:t

    x .- x[end] .* interval ./ t
end # end function

function generate_residuals(Δₜ, t, ξ, ρ)
    B₁ = generate_brownian_bridge(Δₜ, t / 3)
    B₂ = generate_brownian_bridge(Δₜ, t / 3)
    B₃ = generate_brownian_bridge(Δₜ, t / 3)

    vcat(
        ξ .* B₁, 
        ρ .* ξ .* B₁ .+ (ξ .* (1 - ρ ^ 2) ^ .5) .* B₂, 
        ρ .* ξ .* B₂ .+ (ξ .* (1 - ρ ^ 2) ^ .5) .* B₃
    )
end # end function
```

```julia
function calculate_stat(df, length, ℓ)
    Ȳ_global = [sum(Matrix(df[k .+ (0:2) * length, :])) for k in 1:length] / (ℓ * size(df)[2])
    Ȳᵢ = [[sum(df[k .+ (jj - 1) * length, :]) for k in 1:length] / (size(df)[2]) for jj in 1:3]
    Ȳⱼ = [[sum(df[length * (0:ℓ - 1) .+ k, jj]) / ℓ for k in 1:length] for jj in 1:size(df)[2]]
    #[size(df)[2] .* (Ȳᵢ[jj] .- Ȳ_global) .^ 2 for jj in 1:ℓ]
    Ȳᵢ = reduce(hcat, Ȳᵢ)
    SSA = size(df)[2] .* sum((Ȳᵢ .- Ȳ_global) .^ 2, dims = 2)
    SSR = [sum(reduce(hcat, 
          [df[length .* (0:ℓ - 1) .+ k, ii] .- Ȳᵢ[k, ii] .- Ȳⱼ[jj][k] .+ Ȳ_global[k] 
          for jj in 1:size(df)[2] for ii in 1:ℓ]) .^ 2) for k in 1:length]
    F = (SSA / (ℓ - 1)) ./ (SSR / ((ℓ - 1) * (size(df)[2] - 1)))
    
    C = sum(SSA ./ length)
    D = sum(F ./ length)
    E = maximum(F) # <-- this one here

    (C = C, D = D, E = E, F = F, SSA = SSA, SSR = SSR)
end # end function
```

```julia
function permutation_pval(df, length, ℓ, B = 1000)
    stat = calculate_stat(df, length, ℓ)
    n = size(df)[2]
    
    boot_C = Vector{Real}(undef, B)
    boot_D = Vector{Real}(undef, B)
    boot_E = Vector{Real}(undef, B)

    @showprogress .1 "Computing..." for b in 1:B
        df_boot = df
        for jj in 1:n
            for k in 1:length
                indx = k .+ length .* (0:2)
                df_boot[indx, jj] = df_boot[indx[randperm(ℓ)], jj]
            end # end for, k
        end # end for, jj
        stat_boot = calculate_stat(df_boot, length, ℓ)
        x = stat_boot[:C]
        boot_C[b] = stat_boot[:C]
        boot_D[b] = stat_boot[:D]
        boot_E[b] = stat_boot[:E]
    end # end for, b
    (mean(boot_C .> stat[:C]), mean(boot_D .> stat[:D]), mean(boot_E .> stat[:E]))
end # end function
```
